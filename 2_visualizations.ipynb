```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import pandas as pd
```

```python
data_A6002059 = pd.read_csv("datasets/A6002059_gps_ts.csv")
data_A6002058 = pd.read_csv("datasets/A6002058_gps_ts.csv")
data_A7702023 = pd.read_csv("datasets/A7702023_gps_ts.csv")
data_A7702039 = pd.read_csv("datasets/A7702039_gps_ts.csv")
data_A7702043 = pd.read_csv("datasets/A7702043_gps_ts.csv")
data_A7702047 = pd.read_csv("datasets/A7702047_gps_ts.csv")
data_A2302888 = pd.read_csv("datasets/A2302888_gps_ts.csv")
data_A2302900 = pd.read_csv("datasets/A2302900_gps_ts.csv")
data_A2302895 = pd.read_csv("datasets/A2302895_gps_ts.csv")
data_A2302959 = pd.read_csv("datasets/A2302959_gps_ts.csv")
```

```python
all_data = [data_A2302888, data_A2302900, data_A2302895, data_A2302959, data_A6002058,
            data_A6002059, data_A7702023, data_A7702039, data_A7702043, data_A7702047]
```

```python
gps_1 = pd.concat(all_data)
```

# By consumption

```python
def draw_hexbin(df, grid=(120,120)):
    x = df['GpsLongitude']
    y = df['GpsLatitude']
    C = df['FuelConsumption_l_h']
    hx = plt.hexbin(x,y,C,gridsize=grid, cmap='plasma', bins='log')
    cb = plt.colorbar(hx)
    plt.show()
```

```python
plt.figure(figsize=(22,12))
draw_hexbin(gps_1)
```
